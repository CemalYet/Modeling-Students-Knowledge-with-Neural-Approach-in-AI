```python
#STEP-1
#Make Tuple
a=np.array([5,24,16,8,10,6,2,11,10,18,13,10,13,24])
last_id=-1
#14 feaures array
summer_list=np.zeros(14)
#Check if student entered the exam {enter, not_enter-> 0,1 }
bool_lis=np.zeros(14)
#counter for the number of people
person_counter=0
#tuple for the to be  new CSV file
fourteen_feature_tuple=[]
for person in last_df.iterrows():
  #take varialbles
  id,skill,value=person[1]
  if last_id!= -1 and last_id != id:
    person_counter=person_counter+1
    #trashhold 0.5
    summer_list=1*((summer_list/a)>0.5)
    print('id',last_id, 'skills ',summer_list,'isEntered ',bool_lis )
    #add to the list as tuple (person_ID,features,attempted)
    fourteen_feature_tuple.append((last_id,summer_list,bool_lis)) 
    #clear arrays
    summer_list=np.zeros(14)
    bool_lis=np.zeros(14)
    print('---------------------------------------------------------------------------------------------------------------------------------')
  lower=0
  upper=0
  #algortihm for the feature boundries
  for i in range(len(a)):
    upper=upper+a[i]
    if skill >= lower and skill <upper :
      summer_list[i]=summer_list[i]+ value
      bool_lis[i]=1
    lower=upper
  
  last_id=id
#append last person   
person_counter=person_counter+1
summer_list=1*((summer_list/a)>0.5)
print('id',last_id, 'skills ',summer_list,'isEntered ',bool_lis )
fourteen_feature_tuple.append((last_id,summer_list,bool_lis)) 
print('#Person')
print(person_counter)
```

```python
len(fourteen_feature_tuple)
```

```python
#STEP-2
#csv colums
fourteen_id=[]
fourteen_skill_id=[]
fourteen_correct=[]
for fourteen in fourteen_feature_tuple:
  #take values from tuple
  f_id,f_skill,is_entered=fourteen
  for i in range(len(f_skill)):
    #person took the exam then add to list
    if is_entered[i]==1:
      fourteen_id.append(f_id)
      fourteen_skill_id.append(i)
      fourteen_correct.append(f_skill[i])
```

```python
#chechk if the values make sense
print(len(fourteen_id))
print(len(fourteen_skill_id))
print(len(fourteen_correct))
print(fourteen_id[:5])
print(fourteen_skill_id[:5])
print(fourteen_correct[:5])
```

```python
#STEP-3
# turn into dic then pandas 
dict = {'user_id': fourteen_id, 'skill_id': fourteen_skill_id, 'correct': fourteen_correct}  
fourteen_df = pd.DataFrame(dict) 
```
